```python
import pickle
import numpy as np
from pyQus.spectral import SpectralAnalysis
from pyQus.analysisObjects import UltrasoundImage
from src.DataLayer.spectral import SpectralData
from src.DataLayer.dataObjects import ScConfig
from src.Parsers.canonBinParser import findPreset, getImage

def spectral_analysis(image_path, phantom_path, roi_path, config_path):
    imPreset = findPreset(image_path)
    phantomPreset = findPreset(phantom_path)
    assert imPreset == phantomPreset
    
    imgDataStruct, imgInfoStruct, refDataStruct, refInfoStuct = getImage(
        f"{image_path.name}", f"{image_path.parent}", f"{phantom_path.name}", f"{phantom_path.parent}"
    )

    with open(roi_path, "rb") as f:
        roi_info = pickle.load(f)
        
    assert roi_info["Image Name"] == image_path.name
    assert roi_info["Phantom Name"] == phantom_path.name
    
    scSplineX = roi_info["Spline X"]
    scSplineY = roi_info["Spline Y"]
    
    with open(config_path, "rb") as f:
        config_info = pickle.load(f)
    
    config = config_info["Config"]
    
    ultrasoundImage = UltrasoundImage()
    ultrasoundImage.axialResRf = imgInfoStruct.depth / imgDataStruct.rf.shape[0]
    ultrasoundImage.lateralResRf = ultrasoundImage.axialResRf * (
        imgDataStruct.rf.shape[0]/imgDataStruct.rf.shape[1]
    ) # placeholder
    ultrasoundImage.bmode = imgDataStruct.scBmodeStruct.preScArr
    ultrasoundImage.phantomRf = refDataStruct.rf
    ultrasoundImage.rf = imgDataStruct.rf
    ultrasoundImage.scBmode = imgDataStruct.scBmode
    ultrasoundImage.xmap = imgDataStruct.scBmodeStruct.xmap
    ultrasoundImage.ymap = imgDataStruct.scBmodeStruct.ymap
    
    spectralAnalysis = SpectralAnalysis()
    spectralAnalysis.ultrasoundImage = ultrasoundImage
    spectralAnalysis.config = config
    spectralAnalysis.initAnalysisConfig()
    spectralAnalysis.scSplineX = scSplineX
    spectralAnalysis.scSplineY = scSplineY
    
    spectralAnalysis.splineToPreSc()
    spectralAnalysis.generateRoiWindows()
    spectralAnalysis.computeSpecWindows()
    
    scConfig = ScConfig()
    scConfig.width = imgInfoStruct.width1
    scConfig.tilt = imgInfoStruct.tilt1
    scConfig.startDepth = imgInfoStruct.startDepth1
    scConfig.endDepth = imgInfoStruct.endDepth1
    
    spectralData = SpectralData()
    spectralData.spectralAnalysis = spectralAnalysis
    spectralData.scConfig = scConfig
    
    spectralData.convertImagesToRGB()
    spectralData.drawCmaps()
    
    mbfMean = np.mean(spectralData.mbfArr)
    ssMean = np.mean(spectralData.ssArr)
    siMean = np.mean(spectralData.siArr)
    
    return mbfMean, ssMean, siMean
```

```python
analysisFolder = Path("/Volumes/CREST Data/David_S_Data/ATI-Data-CanonFatStudy")
phantom1_path = Path('/Volumes/CREST Data/David_S_Data/ATI-Data-CanonFatStudy/Phantom data/Preset_1/20220831121844_IQ.bin')
phantom2_path = Path('/Volumes/CREST Data/David_S_Data/ATI-Data-CanonFatStudy/Phantom data/Preset_2/20220831121752_IQ.bin')
config_path = Path('/Volumes/CREST Data/David_S_Data/ATI-Data-CanonFatStudy/analysisConfig.pkl')

image_list = []; phantom_list = []; roi_list = []; mbf_list = []
ss_list = []; si_list = []; patient_list = []; image_index_list = []
for patient_folder in analysisFolder.iterdir():
    if patient_folder.is_dir() and not patient_folder.name.startswith('.'):
        try: 
            patient_num = int(patient_folder.name)
            preset = -1
            phantom_path = None
            for preset_folder in patient_folder.iterdir():
                if preset_folder.is_dir() and preset_folder.name == "Preset_1":
                    if len([_ for _ in preset_folder.iterdir()]) > 2:
                        preset = 1
                        phantom_path = phantom1_path
                if preset_folder.is_dir() and preset_folder.name == "Preset_2":
                    if len([_ for _ in preset_folder.iterdir()]) > 2:
                        preset = 2
                        phantom_path = phantom2_path
            if preset != -1:
                for preset_folder in patient_folder.iterdir():
                    if preset_folder.is_dir() and preset_folder.name == "ROIs":
                        i = 0
                        for scan_folder in preset_folder.iterdir():
                            if scan_folder.is_dir() and not scan_folder.name.startswith('.'):
                                for roi in scan_folder.iterdir():
                                    if roi.is_file() and roi.name[-4:] == (".pkl") and not roi.name.startswith('.'):
                                        image_path = patient_folder / f"Preset_{preset}" / f"{scan_folder.name}.bin"
                                        mbf_mean, ss_mean, si_mean = spectral_analysis(
                                            image_path, phantom_path, roi, config_path
                                        )
                                        image_list.append(image_path.name); phantom_list.append(phantom_path.name)
                                        roi_list.append(roi.stem); mbf_list.append(mbf_mean)
                                        ss_list.append(ss_mean); si_list.append(si_mean)
                                        image_index_list.append(i); patient_list.append(patient_num)
                                i += 1                       
        except ValueError:
            pass
```
